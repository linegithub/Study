```python
retornofinanceiro = (virouarray[1:20] - virouarray[0:19]) / virouarray[0:19]
retornofinanceiro
```

```python
t = np.arange(0,19)
t
```

```python
fig.title('Retorno Financeiro')
fig.xlabel('tempo')
fig.ylabel('retorno')
fig.plot(t,retornofinanceiro)
```

```python
import math
import statistics as st
```

```python
media = st.mean(retornofinanceiro)
media
```

```python
desviopadrao = st.pstdev(retornofinanceiro)
desviopadrao
```

```python
n = len(retornofinanceiro)
n
```

```python
otimista = media + ((2*desviopadrao)/math.sqrt(n))
pessimista = media - ((2*desviopadrao)/math.sqrt(n))
print(otimista,pessimista)
```
